```python
import pandas as pd
df_a=pd.read_excel('b1_t_stat_activity_log.xlsx')
df_b=pd.read_excel('b1_t_stat_widget_log.xlsx')
print()

#student attended in dataframe a
activity_counts_a = df_a.groupby(['personid', 'clazzid'])['activity_id'].count().reset_index()
activity_counts_a.columns = ['personid', 'clazzid', 'activity_count_a']

# Count the number of activities each student attended in dataframe b
activity_counts_b = df_b.groupby([ 'clazzid'])['activity_id'].count().reset_index()
activity_counts_b.columns = [ 'clazzid', 'activity_count_b']

print(activity_counts_b)

```

```python
# Merge the two dataframes on 'clazzid'
merged_df = pd.merge(activity_counts_a, activity_counts_b, on=['clazzid'], how='left')

# Fill NaN values with 0 if any
merged_df.fillna(0, inplace=True)

# Print the merged dataframe
print(merged_df)
```

```python
total_activity_count = merged_df.groupby('personid')['activity_count_a','activity_count_b'].sum()
merged_df2=total_activity_count
print(merged_df2)





   
    
    
    
    
```

```python
merged_df2['c'] = merged_df2.apply(lambda row: row['activity_count_a'] / row['activity_count_b'], axis=1)
sorted_df = merged_df2.sort_values(by='c', ascending=False)
sorted_df.head(2000)

sorted_df.to_csv('rate_of_activity.csv')
```
