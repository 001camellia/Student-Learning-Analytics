```python
import pandas as pd
df_a=pd.read_excel('b1_t_stat_exam_answer.xlsx')
score = df_a.groupby(['personid', 'clazzid'])['score'].sum()
average = df_a.groupby(['personid', 'clazzid'])['score'].mean()
std = df_a.groupby(['personid', 'clazzid'])['score'].std()
group_stats = df_a.groupby(['personid', 'clazzid']).agg(
    mean_col1=('score', 'mean'),
    stddev_col1=('score', 'std'),
    max_col2=('score', 'max'),
    min_score=('score', 'min'),
    count=('score','count')
)
#根据平均分排序生成df,如果需要可以另外存储
sorted_df = group_stats.sort_values(by='mean_col1', ascending=False)
#添加考试次数
exam_count = df_a.groupby([ 'clazzid'])['exam_id'].unique()
df = pd.DataFrame(columns=['exam_id'])
df['exam_id'] =exam_count
df['total_exam_number'] = exam_count.transform(lambda x: len(x))
#调整输出格式，给两列添加索引，方便调用
group_stats.index = [(x[0],x[1]) for x in group_stats.index]
b_column = group_stats['clazzid'] = [tup[1] for tup in group_stats.index]
b_column = group_stats['personid'] = [tup[0] for tup in group_stats.index]
#整合所有数据行程df
merged_df = pd.merge(group_stats, df, on=['clazzid'], how='inner')
#调整df的列的order
merged_df = merged_df[['personid','clazzid','mean_col1','stddev_col1','max_col2','min_score','count','total_exam_number']]
```
